```python
"""plotting the Poisson exponential distribution"""
dist = lambda x, c: c*np.exp(-c*x)*(x>0) # using c as my constant as well as some really cool shorthand x>0

xpoints = np.arange(-2, 15, 0.2) # go from -1 to 3 with step size of 0.2
plt.plot(xpoints, dist(xpoints,0.5), 'o')
plt.xlabel('x')
plt.ylabel('c*exp(c*x)')
```

```python
from scipy.stats import expon

x = np.linspace(0, 4, 1000)
colors = sns.cubehelix_palette(4)[::-1]

lambda_ = [0.5, 1, 2 ,4]
plt.figure(figsize=(12,6))
for l,c in zip(lambda_,colors):
    plt.plot(x, dist(x,l), lw=2, color=c, label="$\lambda = %0.1f$"%l)
#    plt.plot(x, expon.pdf(x, scale=1.0/l), lw=2, color=c, label="$\lambda = %0.1f$"%l)
    plt.fill_between(x, dist(x,l), color=c, alpha=0.3)
    
plt.legend()
plt.ylabel("$\lambda e^{-\lambda x}$")
plt.xlabel("$x$")
plt.title("Exponential distribution probability density function (PDF) for various $\lambda$")
```

```python
"""now back to our birth data"""
timediffs = df["minutes after midnight"].diff()[1:] # get the difference in time between adjacent births
timediffs.hist(bins=20)
```

the mean of an exponentially distributed random number is $$\bar{x}=\frac{1}{\lambda}\\$$ and the variance is given by $$\sigma=\frac{1}{\lambda^2}\\$$ therefore the standard deviation is equal to the mean (so long as the data is distributed this way)!

```python
lambda_from_mean = 1.0/timediffs.mean()
print("lam =",lambda_from_mean, "and mean =",1.0/lambda_from_mean)
```

30 min. mean is about what we would expect from 43 births in 24 hours

note that we have been using scipy.stats expon as follows:
"A common parameterization for expon is in terms of the rate parameter lambda, such that pdf = lambda * exp(-lambda * x). This parameterization corresponds to using scale = 1 / lambda."

```python
x = np.linspace(0,160,1000)
plt.plot(x, expon.pdf(x,scale=1.0/lambda_from_mean))
timediffs.hist(normed=True, alpha=0.3, bins=20)
plt.xlabel("Minutes in between adjacent births")
plt.ylabel("Frequency")
plt.title("Normalized data and PDF exponential model")
```

### Aside on the Poisson distribution

probability mass function (Poisson distribution): $$P(k)=\frac{\lambda^k e^{-\lambda}}{k!},~k=0,1,2\ldots\\$$
for data distributed according to this function, lambda appears to be roughly equal to the average as shown below!

```python
from scipy.stats import poisson 

# poisson.pmf(k, mu) = exp(-mu) * mu**k / k! where k=0,1,2,...

# note on enumerate: "enumerate(thing), where thing is either an iterator or a sequence, returns a iterator
#                    that will return (0, thing[0]), (1, thing[1]), (2, thing[2]), and so forth."

k = np.arange(16)
print("k =", k, "and lambda_ =", lambda_) # the underscore after lambda facilitates iteration (no need for [i])
plt.figure(figsize=(12,8))
for i, lambda_ in enumerate([1,2,4,6]):
    plt.plot(k, poisson.pmf(k, lambda_), '-o', label="$\lambda = %0.1f$"%lambda_, color=colors[i])
    plt.fill_between(k, poisson.pmf(k, lambda_), color=colors[i], alpha=0.3)
plt.legend()
plt.xlabel("$k$")
plt.ylabel("$\lambda^ke^{-\lambda}/k!$")
plt.title("Poisson distribution probability mass function")
```

```python
per_hour = df["minutes after midnight"] // 60 # convert to hours (integer)
num_births_per_hour = df.groupby(per_hour)["minutes after midnight"].count() # count 
print(num_births_per_hour)
print("average per bin =", num_births_per_hour.mean())
```

```python
k = np.arange(5)
plt.figure(figsize=(12,8))
tcount = num_births_per_hour.sum()
plt.hist(num_births_per_hour, alpha=0.4, lw=4, normed=True, label="normalized histogram")
sns.kdeplot(num_births_per_hour, label="kernel density estimate")
plt.plot(k, poisson.pmf(k, num_births_per_hour.mean()), '-o', label="probability mass function")
plt.legend()
```

it can be shown that lambda is equal to the mean by maximizing the "likelyhood" function for a given set of exponentially distributed observations:

$$\frac{\partial l(\lambda)}{\partial \lambda}=\frac{\partial}{\partial \lambda}\sum^n_{i=1}\ln(\lambda e^{-\lambda x_i})=0\\$$

we get:

$$\lambda = \frac{1}{n}\sum^n_{i=1}x_i \\$$

## Frequentist statistics

We take a parameter (e.g. lambda) to be fixed and take the data as random, this is the opposite of the Bayesian approach we will see later.  Using the exponential distribution model, we would like to know the lambda for ALL baby births however the best we have is the lambda for our small data set.

Now imagine we M data points representing the means for a bunch of small sets.

```python
from scipy.stats.distributions import bernoulli
def random_up_down(n):
    return bernoulli(.5).rvs(size=n) # rvs stands for 'random variates'
print("e.g.",random_up_down(15))
```

```python
"""return the mean for each sample in an ensamble"""
def acquire_data(number_of_samples, sample_size):
    start = np.zeros((number_of_samples, sample_size), dtype=int)
    for i in range(number_of_samples):
        start[i,:] = random_up_down(sample_size)
    return np.mean(start, axis=1) # get the average from each randonly distributed set
print("e.g.", acquire_data(5,500)) # should return ~0.5 for very large sample sizes
```

```python
"""get 200 means for each sample size from 1 to 1000"""
sample_sizes = np.arange(1,1001,1)
sample_means = [acquire_data(200, i) for i in sample_sizes]
```

```python
mean_of_sample_means = [np.mean(i) for i in sample_means]
```

```python
plt.plot(sample_sizes, mean_of_sample_means)
```

```python
"""using a 'bootstrap' method to re-sample our one data set"""
M_samples = 10000
N_points_per_sample = timediffs.shape[0] # this returns the size of our small data set (43)
bs_np = np.random.choice(timediffs, size=(M_samples, N_points_per_sample))
sd_mean = np.mean(bs_np, axis=1)
sd_std = np.std(bs_np, axis=1)
plt.hist(sd_mean, bins=30, alpha=0.3, normed=True, label='means of random samples')
sns.kdeplot(sd_mean, label='kernel density estimate')
plt.axvline(timediffs.mean(), 0, 1, alpha=0.7, label='our sample')
plt.axvline(timediffs.std()+timediffs.mean(), 0, 1, alpha=0.5, color='r', label='standard deviation of our sample')
plt.axvline(-timediffs.std()+timediffs.mean(), 0, 1, alpha=0.5, color='r')
plt.axvline(sd_mean.std()+sd_mean.mean(), 0, 1, alpha=0.5, color='purple', label='standard deviation of the means')
plt.axvline(-sd_mean.std()+sd_mean.mean(), 0, 1, alpha=0.5, color='purple')
plt.legend()
```

```python
"""using a different 'bootstrap' method (parametric)"""
bs_p = expon(scale=1.0/lambda_from_mean).rvs(size=(M_samples, N_points_per_sample)) # recall that lambda_from_mean~1/30
sd_mean_p = np.mean(bs_p, axis=1)
sd_std_p = np.std(bs_p, axis=1)
plt.hist(sd_mean_p, bins=30, normed=True, alpha=0.3, label='means of random samples')
sns.kdeplot(sd_mean_p, label='kernel density estimate')
plt.legend()
```

A final note on norm=True (from numpy documentation):
