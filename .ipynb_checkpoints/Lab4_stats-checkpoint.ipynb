$$S = \sum{(y_i - (mx_i + b))^2}$$

#### K-fold cross-validation

This concept is an extension of the test/train split. The procedure is:
- randomly assign $n$ samples to one of $K$ groups
- for each group $k$:
    - fit the model on all data excluding that in $k$
    - use the model to predict the outcomes of $k$
    - calculate the prediction error for each observation in the $k$'th group (e.g., $(y_{prediction} - y)^2$)
    - calculate the average prediction error across all samples (e.g., $\frac{1}{n}\sum{(y_{prediction, i} - y_i)^2}$)

#### Do linear regression with `statsmodels`

$y = b + m_1x + \epsilon$ where $y$ is the eruption time and $x$ is the waiting time in between eruptions

```python
x = faithful.waiting
y = faithful.eruptions
model = sm.OLS(y, x)
```

```python
results = model.fit()
results.summary()
```

```python
# Look at the parameters currently fit
results.params.values
```

Notice there is only one parameter (the slope) - we have not fit an intercept $b$. Let's do that now.

```python
x = sm.add_constant(x)
x.head()
```

```python
model_with_int = sm.OLS(y, x)
results_with_int = model_with_int.fit()
results_with_int.params.values
```

```python
print(results.predict(75))
print(results_with_int.predict([1,75]))
```

So if we are waiting for 75 minutes we should expect an erruption to last about 3.8 minutes.

Let's plot the fits below, using `results.fittedvalues` in place of `results.predict(x)`.

```python
plt.scatter(faithful.waiting, faithful.eruptions)
plt.xlabel('Waiting time to next eruption (in mins)')
plt.ylabel('Eruption time (in mins)')
plt.title('Old Faithful Geyser linear models')

plt.plot(faithful.waiting, results.fittedvalues, color='blue',
         linewidth=3, label='Not fitting an intercept ($b=0$)')
plt.plot(faithful.waiting, results_with_int.fittedvalues, color='red',
         linewidth=3, label='Fitting an intercept of $b=1$')
plt.legend(loc='upper left', frameon=True)
```

#### Residuals

Or we can get them like this:

```python
plt.plot(faithful.waiting, resids, 'o')
plt.hlines(y = 0, xmin=40, xmax = 100)
plt.xlabel('Waiting time')
plt.ylabel('Residuals')
plt.title('Residual Plot')
```

#### Use linear regression to predict erruption times with least squares

Instead of using `statsmodel` or `sklearn` let's use the least squares estimates of the coefficients for the linear regression model.

$ \beta = (x^Tx)^{-1}\cdot x^T \cdot y  \equiv A\cdot B \cdot C$

where $\beta = (b, m_1, m_2, ...)$ in general

```python
A = np.dot(x.T, x)
A
```

```python
A = np.linalg.inv(A)
A
```

```python
B = x.T
C = y
```

```python
beta = A.dot(B).dot(C)
beta
```

```python
print("Directly estimating beta:", beta)
print("Estimating beta using statmodels: ", results_with_int.params.values)
```

I've got to say this is really neat, linear algebra is very powerful!

### Logistic regression

There are many different types of regression. Like what if the $y$ variable is discreete or binary? This is what a logistic regression is for.

Take a set of binary observations ($y_1, Y_2, ...$) where each has a set of features $x_1, x_2, ...$. The logistic regression model assumes that there is some set of weights/coefficients/parameters $\beta$ - one for each feature - such that we can generate predictions by flipping a weighted coin. The probability of flipping '1' would be given by:

$$ P(y_i = 1) = \text{logistic}(\sum{\beta_i x_i}) $$

where $\text{logistic}(x) = e^x/(1+e^x)$

#### Our example:

> On January 28, 1986, the space shuttle Challenger exploded and seven astronauts died because two rubber O-rings leaked. These rings had lost their resiliency because the shuttle was launched on a very cold day. Ambient temperatures were in the low 30s and the O-rings themselves were much colder, less than 20F.
One day before the flight, the predicted temperature for the launch was 26F to 29F. Concerned that the rings would not seal at such a cold temperature, the engineers who designed the rocket opposed launching Challenger the next day.

```python
from IPython.display import Image as Im
from IPython.display import display
Im('./Lab4_content/shuttle.png')
```

```python
data=np.array([[float(j) for j in e.strip().split()] for e in open("./Lab4_content/chall.txt")])
data
```

```python
from statsmodels.formula.api import logit, glm, ols

# statsmodels works nicely with pandas dataframes
dat = pd.DataFrame(data, columns = ['Temperature', 'Failure'])
logit_model = logit('Failure ~ Temperature',dat).fit()
logit_model.summary()
```

We are usually looking for a small $p$-value to signify a strong correlation (usually this means we look for $p<0.05$

```python
# calculate predicted failure probabilities for new termperatures
x = np.linspace(50, 85, 1000)
p = logit_model.params
p
```

```python
eta = p['Intercept'] + x*p['Temperature']
y = np.exp(eta)/(1 + np.exp(eta))
```

Let's plot the data along with a range of predicted failure probabilities for unobserved temperatures.

```python
temps, pfail = data[:,0], data[:,1]
plt.scatter(temps, pfail)
axes=plt.gca()
plt.xlabel('Temperature')
plt.ylabel('Failure')
plt.title('O-ring failures')

# plot fitted values
plt.plot(x, y)

# change limits, for a nicer plot
plt.xlim(50, 85)
plt.ylim(-0.1, 1.1)
```
